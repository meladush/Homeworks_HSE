В ячейке ниже - код "определителя душевной боли" от 1 до 10. Давайте его улучшим:

* добавьте условие для случаев, если пользователь вводит число больше 10

* добавьте еще два условия, чтобы разделение на категории душевной боли было еще точнее

```python
a = (input('Оцените силу вашей душевной боли по шкале от 0 до 10 (можно вводить вещественные числа, например 3.6): '))

try:
  a = float(a)
  if a == 0:  # наверное, важно учесть вариант "0"
    print('Кажется, вы счастливый человек! Мы рады за Вас =)')
  if 0 < a < 1.1:
    print('Кажется, вы испытываете легкое огорчение. Как человек в короткой поездке в метро, забывший дома наушники (хотя мы понимаем, что иногда и это может стать большой проблемой).')
  elif 1.1 <= a < 3.2:
    print('Вы встревожены - почти как человек, которому придется звонить по телефону, и это в эпоху мессенджеров! (Хотя мы понимаем, что иногда и это может быть очень тревожной ситуацией)')
  elif 3.2 <= a < 4.2:
    print('Возможно, вы раздражены или огорчены - например как человек, проливший кофе на красивую рубашку. Но может быть, в глобальном масштабе это не так страшно?')
  elif 4.2 <= a < 6.9:
    print('Вы расстроены - например, как человек, у которого что-то не получилось. Пожалуйста, помните, что в большинстве ситуаций всё еще можно исправить!')
  elif 6.9 <= a < 7.9:
    print('Вам грустно или тревожно. Мы понимаем, как может раздражать этот совет, но подумайте, возможно, ничего непоправимого не случилось?')
  elif 7.9 <= a < 9:
    print('Кажется, Вам очень грустно или тревожно. Обнимаем Вас и сочувствуем.')
  elif 9 <= a <= 10:
    print('Вы максимально расстроены или обеспокоены - например как человек, с которым долго не выходит на связь кто-то из близких. Держитесь и постарайтесь найти поддержку.')
  elif a > 10:
    print('''Кажется, Вам очень плохо. Как если бы Вы однажды проснулись в стране, где все, начиная от президента, сошли с ума. Или как если бы Вы потеряли близкого.\
    \nК сожалению, наша шкала не умеет обрабатывать такие данные. Советуем Вам как можно скорее обратиться за психологической помощью.''')

except:
  print('Можно вводить только числа с разделителем "точка". Пожалуйста, проверьте данные.')


```

## Задача 2
### Список, цикл for

Робот, выбирающий пользователю коктейль:

Представьте, что Вы разрабатываете (фрагмент) электронного помощника для ресторана.

У Вас есть список коктейлей, где каждый коктейль представлен в виде списка ингридиентов (у нас получается список списков). Программа рандомным образом должна выбрать один из коктейлей (т.е. один из списков в общем списке) и вывести на экран сообщение: "Сегодня в вашем коктейле будет: " и сам список ингридиентов

Но есть одно условие: в некоторые из списков по ошибке попал "лосось". Строку "лосось" нужно убрать (здесь вам поможет один метод работы со списками) ,но так как лосось есть не везде, нужно предварительно проверить его наличие в списке.

```python
import random

cocktails = [
    ["мартини","грейпфрутовый сок","жасмин","тоник","лосось"],
    ["клубника","какао","мята","марсала"],
    ["водка","томатный сок","лимонный сок","вустерширский соус","черный перец","сельдерей","лосось"],
    ["джин","вермут","ликер мараскино","апельсины","коктейльная вишня","лосось"],
    ["ром","авокадо","сахарный сироп","сливки","лимонный сок","лед"],
    ["красный вермут","тоник","апельсины","лосось"],
    ["только чай"]
        ]

my_list = random.choice(cocktails)
if 'лосось' in my_list:
  my_list.remove('лосось')

print('Сегодня в вашем коктейле будет:')
for i in my_list:
  print(i)
```
