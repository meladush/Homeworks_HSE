```python
# необходимые импорты
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
import random
import re
import time
```

### шаг 1

```python
# соберите все ссылки на страницы с новостями (цикл for + range)
# 1 страница: https://sysblok.ru
# 2 страница: https://sysblok.ru/page/2
# последняя: https://sysblok.ru/page/16

all_page = ['https://sysblok.ru', ]
for i in range(2, 17):
  all_page.append('https://sysblok.ru/page/' + str(i))

all_page
```

### шаг 2

```python
# с каждой страницы соберите ссылки на отдельные новости
# подсказка: родительский тег h2 (проверяйте по нему), нужный нам - a

all_links = []

for i in all_page:
  page = rq.get(i)
  # print(page.status_code)
  soup = BeautifulSoup(page.text, features="html.parser")
  for i in soup.find_all('a'):
    if i.parent.name == 'h2' and i.get('href') not in all_links:
      all_links.append(i.get('href'))

print(len(all_links)) # посмотрим, сколько всего ссылок собралось
all_links[:20] # выведем первые 20
```

### шаг 3

работаем с 1 новостью:

нас интересует название, автор, дата публикации, текст (дополнительно можно спарсить тематические категории)

```python
# соберем отдельные элементы на случайной новости

news0 = all_links[random.randrange(0, len(all_links))]
page0 = rq.get(news0)
soup0 = BeautifulSoup(page0.text, features="html.parser")
print(news0)
```

```python
# для парсинга даты: извлеките текст из тега time
date = soup0.find('time').text
date = pd.to_datetime(date, dayfirst=True)
print(date)
```

```python
# для парсинга заголовка: извлеките текст из тега h1
title = soup0.find('h1').text
title
```

```python
# для парсинга текста: соберите все тексты из тега p и соедините в строку
text = []
text_list = soup0.find_all('p')
for i in text_list:
  if i.text.strip():  # собираем только <p>, внутри которых есть текст
    i = re.sub('\s', ' ', i.text.strip())  # заменяем нестандартные пробелы внутри абзацев на обычный
    text.append(i)
final_text = '\n'.join(text)

final_text
```

```python
# автора собрать сложнее:
# достаем из тега meta как на паре или из тега a с проверкой на содержимое атрибута rel
author = soup0.find('meta', {'name' : 'author'}).attrs['content']
author
```

```python
# тематические категории:
# <a rel="category tag"> нужные нам категории </a>
# можно попробовать так:
# if i.get('rel') == ['category', 'tag']
category = ''
for i in soup0.find_all('a'):
  if i.get('rel') == ['category', 'tag']:
    category = i.text

category
```

### шаг 4

```python
# пишем функцию по сбору информации
def GetNews(url):
  page = rq.get(url)
  soup = BeautifulSoup(page.text, features="html.parser")

  date = soup.find('time').text
  date = pd.to_datetime(date, dayfirst=True)

  title = soup.find('h1').text

  author = soup.find('meta', {'name' : 'author'}).attrs['content']

  text = []
  text_list = soup.find_all('p')
  for i in text_list:
    if i.text.strip():
      i = re.sub('\s', ' ', i.text.strip())
      text.append(i)
  final_text = '\n'.join(text)

  category = ''
  for i in soup0.find_all('a'):
    if i.get('rel') == ['category', 'tag']:
      category = i.text

  return url, date, title, author, category, final_text
```

```python
GetNews(news0)  # проверим функцию на той же случайной новости
```

```python
# собираем все новости!
# полезно использовать try / except

%%time

news = []
parse_failed = []

for link in all_links:
  try:
    new = GetNews(link)
    news.append(new)
  except:
    print(f'Что-то пошло не так на странице: {link}')
    parse_failed.append(link)

with open('failed_parsing.csv', 'w') as file:  # запишем в файл ссылки, которые не спарсились
  file.write('link' + '\n')
  for line in parse_failed:
    file.write(line + '\n')
```

### шаг 5

```python
# соберите всю собранную информацию в датафрейм
df = pd.DataFrame(news)
df.head()
df.columns = ['link', 'date', 'title', 'author', 'category', 'text']
df.to_csv('sysblok.csv')
```
